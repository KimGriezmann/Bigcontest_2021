# HOME_OPS

```python
# Extra Trees

X_home_ops = home_tr.iloc[:, :34].reset_index(drop=True)
y_home_ops = home_tr[['ops_label']].reset_index(drop=True)

train_X, test_X, train_y, test_y = train_test_split(X_home_ops, y_home_ops, test_size = 0.2, random_state = 42)

ET_home_ops = ExtraTreesRegressor(random_state = 42)
ET_home_ops.fit(train_X, train_y)

pred_y = ET_home_ops.predict(test_X)
score = mean_squared_error(test_y, pred_y)
print('Extra Trees : {}'.format(np.sqrt(score)))

# Random Forest

RF_home_ops = RandomForestRegressor(random_state = 42)
RF_home_ops.fit(train_X, train_y)

pred_y = RF_home_ops.predict(test_X)
score = mean_squared_error(test_y, pred_y)
print('Random Forest : {}'.format(np.sqrt(score)))

# LGBM

lgbm_home_ops = LGBMRegressor(random_state=42)
lgbm_home_ops.fit(train_X, train_y)

pred_y = lgbm_home_ops.predict(test_X)
score = mean_squared_error(test_y, pred_y)
print('LGBM : {}'.format(np.sqrt(score)))

# Stacker

stacker_home_ops = StackingRegressor(estimators = [('et', ET_home_ops), ('rf', RF_home_ops), ('lgbm', lgbm_home_ops)])
stacker_home_ops.fit(train_X, train_y)

pred_y = stacker_home_ops.predict(test_X)
score = mean_squared_error(test_y, pred_y)
print('Stacking Model : {}'.format(np.sqrt(score)))
```
