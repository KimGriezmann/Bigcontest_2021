```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hts_2018 = pd.read_csv("/content/2021 빅콘테스트_데이터분석분야_챔피언리그_스포츠테크_HTS_2018.csv", encoding="euc-kr")
hts_2019 = pd.read_csv("/content/2021 빅콘테스트_데이터분석분야_챔피언리그_스포츠테크_HTS_2019.csv", encoding="euc-kr")
hts_2020 = pd.read_csv("/content/2021 빅콘테스트_데이터분석분야_챔피언리그_스포츠테크_HTS_2020.csv", encoding="euc-kr")
hts_2021 = pd.read_csv("/content/2021 빅콘테스트_데이터분석분야_챔피언리그_스포츠테크_HTS_2021.csv", encoding="euc-kr")
hts=pd.concat([hts_2018,hts_2019,hts_2020,hts_2021],ignore_index=True)
```

```python
hts.head()
```

```python
# HIT_RESULT 확인
set(hts['HIT_RESULT'])
```

```python
# HIT_VEL 확인
print("최고 속도 : {0} km/h, \n최저 속도 : {1} km/h".format(max(hts['HIT_VEL']), min(hts['HIT_VEL'])))
plt.figure(figsize=(7,7))
plt.hist(hts['HIT_VEL'], bins=100)
plt.show()
```

```python
# HIT_ANG_VER 확인
print("최고 각도 : {0} 도, \n최저 각도 : {1} 도".format(max(hts['HIT_ANG_VER']), min(hts['HIT_ANG_VER'])))
plt.figure(figsize=(7,7))
plt.hist(hts['HIT_ANG_VER'], bins=100)
plt.show()
```

```python
hit_1 = hts.loc[hts['HIT_RESULT']=='1루타']
hit_2 = hts.loc[hts['HIT_RESULT']=='2루타']
hit_3 = hts.loc[hts['HIT_RESULT']=='3루타']
homerun = hts.loc[hts['HIT_RESULT']=='홈런']
hit_infield = hts.loc[hts['HIT_RESULT']=='내야안타(1루타)']
out_ground = hts.loc[hts['HIT_RESULT']=='땅볼아웃']
out_bunt = hts.loc[hts['HIT_RESULT']=='번트아웃']
hit_bunt = hts.loc[hts['HIT_RESULT']=='번트안타']
out_double = hts.loc[hts['HIT_RESULT']=='병살타']
out_triple = hts.loc[hts['HIT_RESULT']=='삼중살타']
fielder_choice = hts.loc[hts['HIT_RESULT']=='야수선택']
infield_fly = hts.loc[hts['HIT_RESULT']=='인필드플라이']
line_drive = hts.loc[hts['HIT_RESULT']=='직선타']
foul_fly = hts.loc[hts['HIT_RESULT']=='파울플라이']
fly = hts.loc[hts['HIT_RESULT']=='플라이']
sacrifice_bunt = hts.loc[hts['HIT_RESULT']=='희생번트']
sacrifice_fly = hts.loc[hts['HIT_RESULT']=='희생플라이']
```
